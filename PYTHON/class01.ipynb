Python mein Variable Names ke Rules
Variable ka naam chhote letters (a-z), bade letters (A-Z), numbers (0-9) aur underscore (_) par mushtamil ho sakta hai.
Variable ka naam Python ke reserved keywords (jaise if, for, while) nahi ho sakta, lekin in keywords ko kisi lambe naam ka hissa banaya ja sakta hai.
Python mein capital aur small letters alag samjhe jate hain (case-sensitive):
rose aur Rose do mukhtalif variables hain.
Agar rose ko value di jaye to Rose ki value nahi badlegi.

CONCATENATING TEXT STRING

```python
greeting = "Hello, World!"
print(greeting)
```

```python
greeting = "Hello"
addressee = "World"
```

```python
whole_greeting = greeting + addressee
```

```python
print(whole_greeting)
```

```python
greeting = "Hello"
separators = ", "
addressee = "World"
punc = "!"
whole_greeting = greeting + separators + addressee + punc
print(whole_greeting)
```
