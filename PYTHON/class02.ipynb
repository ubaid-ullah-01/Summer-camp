In Python, a data type defines the kind of value a variable can store.

Common Python Data Types
| Data Type | Description                          | Example                      |
| --------- | ------------------------------------ | ---------------------------- |
| `int`     | Whole numbers                        | `10`, `-5`, `1000`           |
| `float`   | Decimal numbers                      | `3.14`, `99.5`               |
| `str`     | Text/String                          | `"Hello"`, `'Python'`        |
| `bool`    | True or False values                 | `True`, `False`              |
| `list`    | Ordered, mutable collection          | `[1, 2, 3]`                  |
| `tuple`   | Ordered, immutable collection        | `(1, 2, 3)`                  |
| `set`     | Unordered collection of unique items | `{1, 2, 3}`                  |
| `dict`    | Key-value pairs                      | `{"name": "Ali", "age": 20}` |


A data type is a classification that tells Python what kind of value a variable stores and what operations can be performed on it.

Simple Definition

A data type tells the computer whether the data is:

- a number,
- text,
- true/false value,
- or a collection of values.

```python
age = 18                          # int
height = 5.9                      # float
name = "Ali"                      # str
is_student = True                 # bool

fruits = ["apple", "banana"]      # list
colors = ("red", "green")         # tuple
numbers = {1, 2, 3}               # set
student = {"name": "Ali", "age": 18}  # dict

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
print(type(fruits))
print(type(colors))
print(type(numbers))
print(type(student))
```

A comparison operator in Python is used to compare two values. It returns either True or False depending on whether the comparison is correct.

- COMMON COMPARISON OPERATOR

| Operator | Meaning                  | Example           |
| -------- | ------------------------ | ----------------- |
| `==`     | Equal to                 | `5 == 5` → `True` |
| `!=`     | Not equal to             | `5 != 3` → `True` |
| `>`      | Greater than             | `10 > 5` → `True` |
| `<`      | Less than                | `3 < 7` → `True`  |
| `>=`     | Greater than or equal to | `5 >= 5` → `True` |
| `<=`     | Less than or equal to    | `4 <= 6` → `True` |


```python
# Equal to (==)
x = 10
print(x == 10)
```

```python
# Not Equal to (!=)
x = 10
print(x != 5)
```

```python
# Greater Than (>)
x = 10
print(x > 5)
```

```python
# Less Than (<)
x = 10
print(x < 20)
```

```python
# Greater Than or Equal To (>=)
x = 10
print(x >= 10)
```

```python
#  Less Than or Equal To (<=)
x = 10
print(x <= 15)
```
