```python
def calc_percentage_no_parameters():
    bio = int(input("enter the marks of bio"))
    chem = int(input("Enter the marks of chem"))
    engl = int(input("enter the number of engl"))
    phy = int(input("Enter the number of phy"))
    
    total_marks = 400
    obtained_marks = bio + chem + engl + phy
    percentage =  (obtained_marks * 100)/total_marks
    print(f"PERCENTAGE IS {percentage}")
```

```python
calc_percentage_no_parameters()
```

```python
def calculating_percentage(bio, chem, engl, phy):#() jo cheez in ka ander sa ay wo parameter hota han or jo bahar sa ay wo arguments
    total_marks = 400
    obtained_marks = bio + chem + engl + phy
    percentage =  (obtained_marks * 100)/total_marks
    print(f"PERCENTAGE IS {percentage}")

    
```

```python
# The number in the brackets are arguments
calculating_percentage(23, 45, 24, 65)
```

- def ka matlab function define karna.
- calculate_volumne function ka naam hai.
- length, breadth, aur height inputs (parameters) hain jo function ko values dene ke liye use hote hain
- return function ka result wapas bhejta hai.
- Is case mein vol ki value return ho rahi hai.

```python
def calculate_volumne(length, breadth, height):
    vol = length * breadth * height
    return vol
```
