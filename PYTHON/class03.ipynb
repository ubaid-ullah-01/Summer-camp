A school has following rules for grading system:
- Below 25 - F

- 25 to 45 - E

- 45 to 50 - D

- 50 to 60 - C

- 60 to 80 - B

- Above 80 - A

Ask user to enter marks and print the corresponding grade.

- Ye code user se marks leta hai aur marks ki range ke mutabiq grade show karta hai.

- elif ka matlab hota hai "else if".

- Ye tab use hota hai jab pehli if condition false ho jaye aur hum doosri condition check karna chahte hon.

    if = agar

    elif = warna agar

    else = warna

```python
marks=int(input("Enter your Marks"))
if marks>=0 and marks<25:
    print(f"your marks  is {marks} and your grade is F")
elif marks >=25 and marks <45:
    print(F"your marks is {marks} and your grade is E")
elif marks >=45 and marks <50:
    print(F"your marks is {marks} and your grade is D")
elif marks >=50 and marks<60:
    print(F"your marks is {marks} and your grade is C")
elif marks >=60 and marks<80:
    print(F"your marks is {marks} and your grade is B")
elif marks >=80 and marks<100:
    print(F"your marks is {marks} and your grade is A")
else:
    print(f"you put an invalid number")
```

A comparison operator in Python is used to compare two values. It returns either True or False depending on whether the comparison is correct.

- COMMON COMPARISON OPERATOR

| Operator | Meaning                  | Example           |
| -------- | ------------------------ | ----------------- |
| `==`     | Equal to                 | `5 == 5` → `True` |
| `!=`     | Not equal to             | `5 != 3` → `True` |
| `>`      | Greater than             | `10 > 5` → `True` |
| `<`      | Less than                | `3 < 7` → `True`  |
| `>=`     | Greater than or equal to | `5 >= 5` → `True` |
| `<=`     | Less than or equal to    | `4 <= 6` → `True` |


```python
# Equal to (==)
x = 10
print(x == 10)
```

```python
# Not Equal to (!=)
x = 10
print(x != 5)
```

```python
# Greater Than (>)
x = 10
print(x > 5)
```

```python
# Less Than (<)
x = 10
print(x < 20)
```

```python
# Greater Than or Equal To (>=)
x = 10
print(x >= 10)
```

```python
#  Less Than or Equal To (<=)
x = 10
print(x <= 15)
```

```python
# Example with if
age = 18

if age >= 18:
    print("You can vote")
```
