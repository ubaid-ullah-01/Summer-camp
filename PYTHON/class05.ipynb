## WHILE LOOP:

A while loop is a programming statement that repeats a block of code as long as a condition is true.

"While something is true, keep doing this task."

while condition:

    # code to repeat

```python
count = 1

while count <= 5:
    print(count)
    count = count + 1
```

How it works
1. count starts at 1.
2. Check: Is count <= 5? → Yes.
3. Print the number.
4. Add 1 to count.
5. Repeat until count becomes 6.
6. 6 <= 5 is false, so the loop stops.

```python
cleanest_cities = ["Cheyenne",
                    "Santa Fe",
                    "Tucson",
                    "Great Falls",
                    "Honolulu"]

user_input = ""
while user_input != "q":
    user_input = input("Enter a city, or q to quit:")
    if user_input != "q":
        for a_clean_city in cleanest_cities:
            if user_input == a_clean_city:
                print("It's one of the cleanest cities")
                break
```
